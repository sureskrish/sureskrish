```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1= pd.read_csv("Advertising.csv")
```

```python
df1.head()
```

```python
x=df1["TV"]
y=df1["radio"]
plt.bar(x,y)
plt.xlabel("TV")
plt.ylabel("radio")
```

```python
x=df1["TV"]
y=df1["radio"]
plt.bar(x,y,width=10,color=["r","b","k"])
plt.xlabel("TV")
plt.ylabel("radio")
```

```python
x=df1["TV"]
y=df1["radio"]
plt.barh(x,y)
plt.xlabel("TV")
plt.ylabel("radio")
```
